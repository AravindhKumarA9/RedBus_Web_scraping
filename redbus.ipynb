```python
# Import necessary libraries
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
import time

# Initialize the Chrome driver
driver = webdriver.Chrome()
driver.maximize_window()
wait = WebDriverWait(driver, 20)

# Define URLs and states
state_urls = [
    {"url": 'https://www.redbus.in/online-booking/apsrtc/?utm_source=rtchometile', "state": "Andhra Pradesh"},
    {"url": 'https://www.redbus.in/online-booking/tsrtc/?utm_source=rtchometile', "state": "Telangana"},
    {"url": 'https://www.redbus.in/online-booking/hrtc/?utm_source=rtchometile', "state": "Himachal Pradesh"},
    {"url": 'https://www.redbus.in/online-booking/ktcl/?utm_source=rtchometile', "state": "Goa"},
    {"url": 'https://www.redbus.in/online-booking/astc/?utm_source=rtchometile', "state": "Assam"},
    {"url": 'https://www.redbus.in/online-booking/rsrtc/?utm_source=rtchometile', "state": "Rajasthan"},
    {"url": 'https://www.redbus.in/online-booking/sikkim-nationalised-transport-snt', "state": "Sikkim"},
    {"url": 'https://www.redbus.in/online-booking/meghalaya-transport-corporation-mtc', "state": "Meghalaya"},
    {"url": 'https://www.redbus.in/online-booking/ksrtc-kerala/?utm_source=rtchometile', "state": "Kerala"},
    {"url": 'https://www.redbus.in/online-booking/chandigarh-transport-undertaking-ctu', "state": "Chandigarh"}
]

route_data_list = []
bus_info_list = []
processed_buses = set()

# Function to scroll down the page:
def scroll_down():
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(5)

# Scrape route names and links on the current page
def route_data(state_name):
    route_container = wait.until(EC.presence_of_all_elements_located((By.XPATH, "//div[@class='route_link']")))
    time.sleep(5)
    for route in route_container:
        route_name = route.find_element(By.XPATH, ".//a[@class='route']").text
        route_link = route.find_element(By.XPATH, ".//a[@class='route']").get_attribute('href')
        route_data_list.append({
            "state_name": state_name,
            "route_name": route_name,
            "route_link": route_link
        })
    scroll_down()
    time.sleep(5)

# Pagination to scrape data from each page
def paginate_route_data(state_name):
    page_number = 1
    while True:
        try:
            route_data(state_name)
            pagination_container = wait.until(EC.presence_of_element_located((By.XPATH, '//*[@id="root"]/div/div[4]/div[12]')))
            next_page_button = pagination_container.find_element(By.XPATH, f'.//div[contains(@class, "DC_117_pageTabs") and text()="{page_number + 1}"]')
            driver.execute_script("arguments[0].scrollIntoView();", next_page_button)
            time.sleep(5)
            next_page_button.click()
            page_number += 1
        except Exception as e:
            break  # Break loop if error or no more pages

# Scrape bus details on each link
def bus_details(route_name, route_link, state_name):
    bus_items = driver.find_elements(By.XPATH, "//div[@class='clearfix bus-item']")
    for bus in bus_items:
        try:
            bus_name = bus.find_element(By.XPATH, ".//div[@class='travels lh-24 f-bold d-color']").text
        except:
            bus_name = "N/A"
        try:
            departing_time = bus.find_element(By.XPATH, ".//div[@class='dp-time f-19 d-color f-bold']").text
        except:
            departing_time = "N/A"
        
        bus_identifier = (bus_name, departing_time)
        if bus_identifier in processed_buses:
            continue  # Skip if bus is already processed
        
        try:
            bus_type = bus.find_element(By.XPATH, ".//div[@class='bus-type f-12 m-top-16 l-color evBus']").text
        except:
            bus_type = "N/A"
        try:
            duration = bus.find_element(By.XPATH, ".//div[@class='dur l-color lh-24']").text
        except:
            duration = "N/A"
        try:
            arrival_time = bus.find_element(By.XPATH, ".//div[@class='bp-time f-19 d-color disp-Inline']").text
        except:
            arrival_time = "N/A"
        try:
            star_rating = bus.find_element(By.XPATH, ".//div[@class='rating-sec lh-24']").text
        except:
            star_rating = "N/A"
        try:
            fare_price = bus.find_element(By.XPATH, ".//span[@class='f-bold f-19']").text
        except:
            try:
                fare_price = bus.find_element(By.XPATH, ".//span[@class='f-19 f-bold']").text
            except:
                fare_price = "N/A"
        try:
            seat_availablity = bus.find_element(By.XPATH, ".//div[@class='seat-left m-top-30']").get_attribute('textContent').split()[0]
        except:
            try:
                seat_availablity = bus.find_element(By.XPATH, ".//div[@class='seat-left m-top-16']").get_attribute('textContent').split()[0]
            except:
                seat_availablity = "N/A"
        
        bus_info_list.append({
            "state_name": state_name,
            "route_name": route_name,
            "route_link": route_link,
            "bus_name": bus_name,
            "bus_type": bus_type,
            "departing_time": departing_time,
            "duration": duration,
            "arrival_time": arrival_time,
            "star_rating": star_rating,
            "fare_price": fare_price,
            "seat_availablity": seat_availablity
        })
        processed_buses.add(bus_identifier)

# Keep scrolling to load all bus details
def scrape_all_bus_details(route_name, route_link, state_name):
    last_height = driver.execute_script("return document.body.scrollHeight")
    while True:
        bus_details(route_name, route_link, state_name)
        scroll_down()
        time.sleep(5)
        new_height = driver.execute_script("return document.body.scrollHeight")
        if new_height == last_height:
            break
        last_height = new_height

# Click the "View Buses" button
def view_bus_button():
    view_bus_button = wait.until(EC.element_to_be_clickable((By.XPATH, "//div[@class='button' and text()='View Buses']")))
    view_bus_button.click()
    time.sleep(5)

# Iterate through each route link to scrape bus details
def scrape_bus_details(route_name, route_link, state_name):
    driver.get(route_link)
    time.sleep(5)
    try:
        view_bus_button()
    except:
        print(f"'View Buses' button not found for {route_name}")
    scrape_all_bus_details(route_name, route_link, state_name)

# Main loop for each state
for state in state_urls:
    driver.get(state["url"])
    time.sleep(5)
    paginate_route_data(state["state"])

# Iterate through route data and scrape bus details
for route in route_data_list:
    try:
        scrape_bus_details(route["route_name"], route["route_link"], route["state_name"])
    except Exception as e:
        continue

# Save the data to a CSV file
bus_details_df = pd.DataFrame(bus_info_list)
bus_details_df.to_csv('bus_statistics_information.csv', index=False)

# Close the browser
driver.quit()
```

```python
import numpy as np

def create_table_from_dataframe(df, table_name, connection):
    # Mapping data types
    dtype_mapping = {
        'int64': 'INT',
        'float64': 'FLOAT',
        'object': 'TEXT',
        'datetime64[ns]': 'DATETIME'
    }
    
    # Generate column definitions for the SQL CREATE TABLE statement
    columns = ', '.join(
        f"{col} {dtype_mapping.get(str(dtype), 'TEXT')}" 
        for col, dtype in zip(df.columns, df.dtypes)
    )
    
    # SQL query to create the table
    create_table_sql = f"CREATE TABLE IF NOT EXISTS {table_name} ({columns});"
    
    # Execute the SQL query
    with connection.cursor() as cursor:
        cursor.execute(create_table_sql)
        connection.commit()

def insert_dataframe_to_sql(df, table_name, connection):
    # Ensure all data types are converted to native Python types
    df = df.astype({
        col: 'int' if dtype == np.int64 else 'float' if dtype == np.float64 else 'str' 
        for col, dtype in zip(df.columns, df.dtypes)
    })
    
    # Create a SQL query to insert values into the table
    sql = f"INSERT INTO {table_name} VALUES ({', '.join(['%s'] * len(df.columns))})"
    
    # Convert DataFrame rows to tuples of native Python types
    data = [tuple(row) for row in df.itertuples(index=False, name=None)]
    
    # Insert the data into the table
    with connection.cursor() as cursor:
        cursor.executemany(sql, data)
        connection.commit()

# Create tables and insert data
create_table_from_dataframe(Cleaned_redbus_df, "Redbus", myconnection)
insert_dataframe_to_sql(Cleaned_redbus_df, "Redbus", myconnection)
```
